점선 그래프 그리기

```python
data = np.random.randn(50).cumsum()
data
```

```python
plt.plot(data) # 데이터로 그래프 그리는 함수
plt.show() # 그래프 보여주는 함수
```

여러 그래프 그릴 준비

```python
plt.subplot(1,2,1) #첫번째 행, 두번째 열의 첫번째 그래프
plt.subplot(1,2,2) #첫번째 행, 두번째 열의 두번째 그래프
plt.show()
```

```python
plt.subplot(2, 3, 1)
plt.subplot(2, 3, 2)
plt.subplot(2, 3, 3)
plt.subplot(2, 3, 4)
plt.subplot(2, 3, 5)
plt.subplot(2, 3, 6)
plt.show()
```

여러 그래프 그리기


```python
hist_data = np.random.randn(100)
scat_data = np.arange(30)
```

```python
plt.subplot(2,2,1)
plt.plot(data)
#그래프
plt.subplot(2,2,2)
plt.hist(hist_data, bins = 20)
# 히스토그램
plt.subplot(2,2,3)
plt.scatter(scat_data, np.arange(30) + 3* np.random.randn(30))
#분산형
plt.show()
```

그래프 선 옵션

그래프를 그릴 때 표시되는 색이나 마커 패턴을 바꾸는 것
색상: b, g, r, c, y,k,w
마커: o(원), v(역삼각형), ^(삼각형), s(네모), +(플러스), .(점)

```python
plt.plot(data, 'gv')
plt.show()
```

```python
plt.plot(data, 'k+')
plt.show()
```

그래프 크기 조절
plt.figure안에 figsize를 이용하여 가로, 세로 길이 조절 가능(인치 단위..)

```python
plt.figure(figsize=(6, 6))
plt.plot(data)
plt.show()
```

여러 그래프 크기 조절

```python
plt.figure(figsize=(8,8)) #맨 위에 있으면 밑의 그래프 크기 전체 조절

plt.subplot(2,2,1)
plt.plot(data)
#그래프
plt.subplot(2,2,2)
plt.hist(hist_data, bins = 20)
# 히스토그램
plt.subplot(2,2,3)
plt.scatter(scat_data, np.arange(30) + 3* np.random.randn(30))
#분산형
plt.show()
```

비율 조절

그래프 겹치기+ legend 달기

```python
data = np.random.randn(30).cumsum()

plt.plot(data, 'k--', label = 'Default')
plt.plot(data, 'k-', drawstyle = 'steps-post', label = 'steps-post')
plt.legend()
plt.show()
```

이름 달기

```python
plt.plot(np.random.randn(1000).cumsum())
plt.title('Random Graph')
plt.xlabel('stages')
plt.ylabel('values')
plt.show()
```

종합

```python
plt.plot(np.random.randn(1000).cumsum(), 'k^', label='one')
plt.plot(np.random.randn(1000).cumsum(), 'b.', label='two')
plt.plot(np.random.randn(1000).cumsum(), 'r', label='three')
plt.legend()
plt.show
```
