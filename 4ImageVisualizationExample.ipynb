이미지 시각화

package 불러오기

```python
import numpy as np
from PIL import Image

import matplotlib.pyplot as plt

%matplotlib inline
```

그림 파일 열기

```python
path = 'source/aqu.PNG'

image_pil = Image.open(path)
image = np.array(image_pil)
```

```python
image.shape
```

```python
np.min(image), np.max(image)
```

그래프로 시각화하기

```python
plt.hist(image.ravel(), 256, [0, 256])
plt.show()
```

```python
plt.imshow(image)
plt.show()
```

그림을 흑백으로 열기

```python
path
```

```python
image_pil = Image.open(path).convert("L")
image_bw = np.array(image_pil)
```

```python
image_bw.shape
```

```python
plt.imshow(image_bw, 'gray')
plt.show()
```

다른 색상으로 cmap 표현하기
gray scale

```python
image_bw.shape
```

jet

```python
plt.imshow(image_bw, 'jet')
plt.show()
```

```python
plt.imshow(image_bw, 'jet')
plt.colorbar()
plt.show()
```

그림 크기 조절

```python
plt.figure(figsize =(10,10))
plt.imshow(image)
plt.show()
```

그림에 제목 추가

```python
plt.title('aqua')
plt.imshow(image)
plt.show()
```
